```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Install necessary libraries
!pip install pydicom numpy monai

import pydicom
import numpy as np
import os

# Function to crop DICOM images
def crop_dicom(image_path, crop_box):
    print(f"Cropping file: {image_path}")
    # Read the DICOM file
    dataset = pydicom.dcmread(image_path)

    # Extract the image data
    image = dataset.pixel_array

    # Perform the cropping
    cropped_image = image[crop_box[1]:crop_box[3], crop_box[0]:crop_box[2]]

    # Update the dataset with the cropped image
    dataset.PixelData = cropped_image.tobytes()
    dataset.Rows, dataset.Columns = cropped_image.shape

    return dataset

# Define the crop box (left, upper, right, lower)
crop_box = (50, 50, 450, 450)

# Path to the input and output directories in Google Drive
input_dir = '/content/drive/My Drive/manifest/NLST/100002/'  # Base input directory on Google Drive
output_dir = '/content/drive/My Drive/cropped_dicom_files/'
os.makedirs(output_dir, exist_ok=True)

# Function to process each DICOM file in a nested directory structure
def process_dicom_files(base_dir, output_dir, crop_box):
    file_count = 0
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if file.endswith('.dcm'):
                input_path = os.path.join(root, file)
                relative_path = os.path.relpath(input_path, base_dir)
                output_path = os.path.join(output_dir, relative_path)
                os.makedirs(os.path.dirname(output_path), exist_ok=True)

                # Crop the DICOM file
                cropped_dataset = crop_dicom(input_path, crop_box)

                # Save the cropped DICOM file
                cropped_dataset.save_as(output_path)

                file_count += 1
                print(f"Saved cropped file: {output_path}")

    if file_count == 0:
        print("No DICOM files found.")
    else:
        print(f"Processed {file_count} DICOM files.")

# Process the DICOM files
process_dicom_files(input_dir, output_dir, crop_box)

print("Cropping completed!")
```
